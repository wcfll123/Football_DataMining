## K-means clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### 2008

```python
# df_2008_cluster_features = df_2008.loc[:, ['crossing','finishing','heading_accuracy',
#                                            'short_passing','volleys','dribbling','curve',
#                                            'free_kick_accuracy','long_passing','ball_control',
#                                            'acceleration','sprint_speed','agility','reactions',
#                                            'balance','shot_power','jumping','stamina','strength',
#                                            'long_shots','aggression','interceptions','positioning',
#                                            'vision','penalties','marking','standing_tackle','sliding_tackle']]
# df_2008_cluster_features
df_2008_cluster_features = df_2008.loc[:, ['crossing','finishing','short_passing','volleys','dribbling','curve',
                                           'long_passing','ball_control','acceleration','agility',
                                           'shot_power','jumping','stamina','strength','long_shots',
                                           'interceptions','standing_tackle','sliding_tackle']]
df_2008_cluster_features
```

```python
numerical_columns = df_2008_cluster_features.columns
X = df_2008_cluster_features[numerical_columns]
```

```python
print(X.isna().sum())
```

```python
X = X.dropna()
```

```python
print(X.isna().sum())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
n_clusters_num = 4
kmeans = KMeans(n_clusters=n_clusters_num, random_state=0)
cluster_labels = kmeans.fit_predict(X_scaled)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
silhouette_avg = silhouette_score(X_scaled, cluster_labels)
print("silhouette_coef：", silhouette_avg)
```

```python
print("cluster_center：\n", kmeans.cluster_centers_)
```

```python
sample_silhouette_values = silhouette_samples(X, cluster_labels)
sample_silhouette_values
```

```python
y_lower = 10
for i in range(n_clusters_num):
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = cm.nipy_spectral(float(i) / n_clusters_num)
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values, facecolor=color, edgecolor=color, alpha=0.7)

    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    y_lower = y_upper + 10

plt.title("The silhouette plot for the various clusters.")
plt.xlabel("The silhouette coefficient values")
plt.ylabel("Cluster label")
plt.axvline(x=silhouette_avg, color="red", linestyle="--")

plt.yticks([])
plt.xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

plt.show()
```

```python
plt.figure(figsize=(8, 6))
for i in range(kmeans.n_clusters):
    plt.scatter(X_pca[cluster_labels == i, 0], X_pca[cluster_labels == i, 1], label=f'Cluster {i}')

centers = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.75, marker='x', label='Centers')

plt.title('KMeans Clustering with PCA')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.legend()
plt.show()
```

### DTW (Dynamic Time Warping)

```python
# !pip install dtaidistance
```

```python
from scipy.spatial.distance import euclidean
from dtaidistance import dtw
```

```python
def calc_pairwise_dtw_cost(x, y, ret_matrix=True):
    DTW = np.zeros((len(x), len(y)))
    for i in range(len(x)):
        for j in range(len(y)):
            if i == 0:
                DTW[i][j] = (x[i]-y[j])**2 + DTW[i][j-1]
            elif j == 0:
                DTW[i][j] = (x[i]-y[j])**2 + DTW[i-1][j]
            else:
                DTW[i][j] = (x[i]-y[j])**2 + min(DTW[i][j-1], \
                                                 DTW[i-1][j], \
                                                 DTW[i-1][j-1])
    if ret_matrix:
        return DTW
    else:
        return DTW[len(x)-1][len(y)-1]
```

```python
calc_pairwise_dtw_cost([1,2,3,4,5], [3,2,5,4,5])
```

```python
def dtw_distance(series1, series2):
    n, m = len(series1), len(series2)
    dtw_matrix = np.zeros((n+1, m+1))

    dtw_matrix[0, :] = np.inf
    dtw_matrix[:, 0] = np.inf
    dtw_matrix[0, 0] = 0

    for i in range(1, n+1):
        for j in range(1, m+1):
            cost = abs(series1[i-1] - series2[j-1])
            last_min = min(dtw_matrix[i-1, j], dtw_matrix[i, j-1], dtw_matrix[i-1, j-1])
            dtw_matrix[i, j] = cost + last_min

    return dtw_matrix[n, m]
```

```python
dtw_results = {}

countries = spec_country_short_passing.columns

for i in range(len(countries)):
    for j in range(i+1, len(countries)):
        country1, country2 = countries[i], countries[j]
        distance = dtw_distance(spec_country_short_passing[country1].values, \
                                spec_country_short_passing[country2].values)
        dtw_results[(country1, country2)] = distance

for pair, distance in dtw_results.items():
    print(f"DTW distance between {pair[0]} and {pair[1]}: {distance}")
```

```python
dtw_results_list = [(i, j) for i, j in dtw_results.items()]

dtw_results_sorted = sorted(dtw_results_list, key=lambda x: x[1])

top_5_dtw_pairs = dtw_results_sorted[1:10]

for pair, distance in top_5_dtw_pairs:
    print(f"DTW distance between {pair[0]} and {pair[1]}: {distance}")
```

Most countries have a high similarity in the trend of changes in the number of passes, with the highest being Germany and Poland, possibly influenced by the popular tactics of that time: the tiki-taka of Spain, the 2010 World Cup champions.
